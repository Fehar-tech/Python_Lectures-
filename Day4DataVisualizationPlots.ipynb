### Type of Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Set the style to whitegrid
sns.set(style="whitegrid")

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Define the order for clarity ranking from highest to lowest
clarity_ranking = ['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1']

# Convert clarity to categorical type with the defined ranking order
diamonds['clarity'] = pd.Categorical(diamonds['clarity'], categories=clarity_ranking, ordered=True)

# Set the plot size
plt.figure(figsize=(10, 6))

# Create a scatter plot with carat and price, and use clarity for color
sns.scatterplot(x="carat", y="price", data=diamonds, hue="clarity", palette="coolwarm", alpha=0.7, size="carat", sizes=(20, 200))

# Despine the plot
sns.despine(left=True, bottom=False)

# Position the legend at the bottom
plt.legend(bbox_to_anchor=(0.5, -0.1), loc='upper center', ncol=6)

# Show the plot
plt.show()
```

# Strip Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style to whitegrid
sns.set(style="whitegrid")

# Load the planets dataset
planets = sns.load_dataset("planets")

# Set the plot size
plt.figure(figsize=(12, 6))

# Create a strip plot with customizations
sns.stripplot(x="method", y="mass", data=planets, 
              jitter=True, 
              palette="Paired", 
              alpha=0.6, 
              size=5, 
              edgecolor='black', 
              linewidth=0.5)

# Customize the plot
plt.title("Distribution of Exoplanet Mass by Discovery Method", fontsize=16)
plt.xlabel("Discovery Method", fontsize=14)
plt.ylabel("Mass (Jupiter Masses)", fontsize=14)
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.grid(True, linestyle='--', alpha=0.7)  # Add gridlines

# Show the plot
plt.tight_layout()  # Adjust layout for better fitting
plt.show()

```

# Corelational Plot


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Select only numeric columns for correlation
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_tips.corr()

# Set the plot size
plt.figure(figsize=(10, 8))

# Create a heatmap for the correlation matrix
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True, cbar_kws={"shrink": .8})

# Customize the plot
plt.title("Correlation Heatmap of Tips Dataset", fontsize=16)

# Show the plot
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Select only numeric columns for correlation
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])

# Check if there are numeric columns
if numeric_tips.empty:
    raise ValueError("No numeric columns available for correlation.")

# Calculate the correlation matrix
correlation_matrix = numeric_tips.corr()

# Set the plot size
plt.figure(figsize=(10, 8))

# Create a heatmap for the correlation matrix
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True, cbar_kws={"shrink": .8})

# Customize the plot
plt.title("Correlation Heatmap of Tips Dataset", fontsize=16)
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.yticks(rotation=0)    # Rotate y labels for better readability

# Show the plot
plt.tight_layout()  # Adjust layout for better fitting
plt.show()


```

# Projection Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
iris = load_iris()
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_data['species'] = iris.target

# Perform PCA
pca = PCA(n_components=2)
iris_pca = pca.fit_transform(iris.data)

# Create a DataFrame for the PCA results
pca_df = pd.DataFrame(data=iris_pca, columns=['PCA1', 'PCA2'])
pca_df['species'] = iris.target

# Create a projection plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='species', data=pca_df, palette='Set2', markers=['o', 's', 'D'])

# Customize the plot
plt.title("PCA Projection Plot of Iris Dataset", fontsize=16)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

# Show the plot
plt.show()
```
